## How to use this tool
**Steps: **
    - Copy all report excel files into the same directory as the tool
    - Click "run'
    - Look for output.csv in the directory
    
**Note:**

Each sheet in the Excel files must contain a report, otherwise, if a blank sheet is present it will cause an error
All excel files must be in the ".xls" format
This tool was written by Brian Mc Donald bmcdonald@unicef.org

```python
import pandas as pd
import os
import glob
dftemplate = pd.DataFrame([['','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','','',0]], columns = ['Woreda_P-Code','Year','Month','Region','Zone','Woreda','Name_OTP/SC','Prog_Type','MHNT','Nb_OTP','Nb_TFU','Impl_Partner','Total_Nb_TFP','Reporting_Rate','BEGML6','BEGM6-69','BEGY5-10','BEGY11-17','BEGYG18','Total_Beginning_of_Month','WH-MUACML6','WH-MUACM6-59','WH-MUACY5-10','WH-MUACY11-17','WH-MUACYG18','Total_WH-MUAC','OEDML6','OEDM6-59','OEDY5-10','OEDY11-17','OEDYG18','Total_Oedema','RELML6','RELM6-59','RELY5-10','RELY11-17','RELYG18','Total_Relapse','RADML6','RADM6-59','RADY5-10','RADY11-17','RADYG18','Total_Readmission','TRIML6','TRIM6-59','TRIY5-10','TRIY11-17','TRIYG18','Total_Transfer-in','TADMML6','TADMM6-59','TADMY5-10','TADMY11-17','TADMYG18','Total_Admission','CURML6','CURM6-59','CURY5-10','CURY11-17','CURYG18','Total_Cured','DIEML6','DIEM6-59','DIEY5-10','DIEY11-17','DIEYG18','Total_Death','UNKML6','UNKM6-59','UNKY5-10','UNKY11-17','UNKYG18','Total_Unknown','DEFML6','DEFM6-59','DEFY5-10','DEFY11-17','DEFYG18','Total_Defaulter','NRML6','NRM6-59','NRY5-10','NRY11-17','NRYG18','Total_Non-responder','MTML6','MTM6-59','MTY5-10','MTY11-17','MTYG18','Total_Medical_Transfer','TOPML6','TOPM6-59','TOPY5-10','TOPY11-17','TOPYG18','Total_Out_to_OTP','TIPML6','TIPM6-59','TIPY5-10','TIPY11-17','TIPYG18','Total_Out_to_IPD','DISML6','DISM6-59','DISY5-10','DISY11-17','DISYG18','Total_Discharge','ENDML6','ENDM6-59','ENDY5-10','ENDY11-17','ENDYG18','Total_End_of_Month'])
df = dftemplate
df.dropna()
df.to_csv('output.csv')
# os.chdir('/Users/brianmcdonald/Sync/Data & Tools/TFP database data-entry tool/')
FileList = glob.glob('*.xls')
for f in FileList:
    dffile = pd.ExcelFile(f)
    file = dffile.sheet_names
    
    for sheet in file:
        dfinput = dffile.parse(sheet)

        df['Name_OTP/SC']=dfinput.iat[1,2]
        df['Region']=dfinput.iat[2,2]
        df['Zone']=dfinput.iat[3,2]
        df['Woreda']=dfinput.iat[4,2]

        df['Prog_Type'] = str(dfinput.iat[4,12]) + str(dfinput.iat[4,14])
        df['MHNT'] = str(dfinput.iat[4,16])

        df['BEGML6'] = dfinput.iat[10,1]
        df['BEGM6-69'] = dfinput.iat[11,1]
        df['BEGY5-10'] = dfinput.iat[12,1]
        df['BEGY11-17'] = dfinput.iat[13,1]
        df['BEGYG18'] = dfinput.iat[14,1]
        df['Total_Beginning_of_Month'] = df['BEGML6'] + df['BEGM6-69'] + df['BEGY5-10']+df['BEGY11-17']+df['BEGYG18']

        df['WH-MUACML6'] = dfinput.iat[10,2]
        df['WH-MUACM6-59'] = dfinput.iat[11,2]
        df['WH-MUACY5-10'] = dfinput.iat[12,2]
        df['WH-MUACY11-17'] = dfinput.iat[13,2]
        df['WH-MUACYG18'] = dfinput.iat[14,2]
        df['Total_WH-MUAC'] = df['WH-MUACML6']+df['WH-MUACM6-59']+df['WH-MUACY5-10']+df['WH-MUACY11-17']+df['WH-MUACYG18']

        df['OEDML6'] = dfinput.iat[10,3]
        df['OEDM6-59'] = dfinput.iat[11,3]
        df['OEDY5-10'] = dfinput.iat[12,3]
        df['OEDY11-17'] = dfinput.iat[13,3]
        df['OEDYG18'] = dfinput.iat[14,3]
        df['Total_Oedema'] = df['OEDML6']+df['OEDM6-59']+df['OEDY5-10']+df['OEDY11-17']+df['OEDYG18']

        df['RELML6'] = dfinput.iat[10,4]
        df['RELM6-59'] = dfinput.iat[11,4]
        df['RELY5-10'] = dfinput.iat[12,4]
        df['RELY11-17'] = dfinput.iat[13,4]
        df['RELYG18'] = dfinput.iat[14,4]
        df['Total_Relapse'] = df['RELML6']+df['RELM6-59']+df['RELY5-10']+df['RELY11-17']+df['RELYG18']

        df['RADML6'] = dfinput.iat[10,5]
        df['RADM6-59'] = dfinput.iat[11,5]
        df['RADY5-10'] = dfinput.iat[12,5]
        df['RADY11-17'] = dfinput.iat[13,5]
        df['RADYG18'] = dfinput.iat[14,5]
        df['Total_Readmission'] = df['RADML6']+df['RADM6-59']+df['RADY5-10']+df['RADY11-17']+df['RADYG18']

        df['TRIML6'] = dfinput.iat[10,6]
        df['TRIM6-59'] = dfinput.iat[11,6]
        df['TRIY5-10'] = dfinput.iat[12,6]
        df['TRIY11-17'] = dfinput.iat[13,6]
        df['TRIYG18'] = dfinput.iat[14,6]
        df['Total_Transfer-in'] = df['TRIML6']+df['TRIM6-59']+df['TRIY5-10']+df['TRIY5-10']+df['TRIY11-17']+df['TRIYG18']

        df['TADMML6'] = df['WH-MUACML6']+df['OEDML6']+df['RELML6']+df['RADML6']+df['TRIML6']
        df['TADMM6-59'] = df['WH-MUACM6-59']+df['OEDM6-59']+df['RELM6-59']+df['RADM6-59']+df['TRIM6-59']
        df['TADMY5-10'] = df['WH-MUACY5-10']+df['OEDY5-10']+df['RELY5-10']+df['RADY5-10']+df['TRIY5-10']
        df['TADMY11-17'] = df['WH-MUACY11-17']+df['OEDY11-17']+df['RELY11-17']+df['RADY11-17']+df['TRIY11-17']
        df['TADMYG18'] = df['WH-MUACYG18']+df['OEDYG18']+df['RELYG18']+df['RADYG18']+df['TRIYG18']
        df['Total_Admission'] = df['Total_WH-MUAC']+df['Total_Oedema']+df['Total_Relapse']+df['Total_Readmission']+df['Total_Readmission']+df['Total_Transfer-in']

        df['CURML6'] = dfinput.iat[10,8]
        df['CURM6-59'] = dfinput.iat[11,8]
        df['CURY5-10'] = dfinput.iat[12,8]
        df['CURY11-17'] = dfinput.iat[13,8]
        df['CURYG18'] = dfinput.iat[14,8]
        df['Total_Cured'] = df['CURML6']+df['CURM6-59']+df['CURY5-10']+df['CURY11-17']+df['CURYG18']

        df['DIEML6'] = dfinput.iat[10,9]
        df['DIEM6-59'] = dfinput.iat[11,9]
        df['DIEY5-10'] = dfinput.iat[12,9]
        df['DIEY11-17'] = dfinput.iat[13,9]
        df['DIEYG18'] = dfinput.iat[14,9]
        df['Total_Death'] = df['DIEML6']+df['DIEM6-59']+df['DIEY5-10']+df['DIEY5-10']+df['DIEY11-17']+df['DIEYG18']

        df['UNKML6'] = dfinput.iat[10,10]
        df['UNKM6-59'] = dfinput.iat[11,10]
        df['UNKY5-10'] = dfinput.iat[12,10]
        df['UNKY11-17'] = dfinput.iat[13,10]
        df['UNKYG18'] = dfinput.iat[14,10]
        df['Total_Unknown'] = df['UNKML6']+df['UNKM6-59']+df['UNKY5-10']+df['UNKY11-17']+df['UNKYG18']

        df['DEFML6'] = dfinput.iat[10,11]
        df['DEFM6-59'] = dfinput.iat[11,11]
        df['DEFY5-10'] = dfinput.iat[12,11]
        df['DEFY11-17'] = dfinput.iat[13,11]
        df['DEFYG18'] = dfinput.iat[14,11]
        df['Total_Defaulter'] = df['DEFML6']+df['DEFM6-59']+df['DEFY5-10']+df['DEFY11-17']+df['DEFYG18']

        df['NRML6'] = dfinput.iat[10,12]
        df['NRM6-59'] = dfinput.iat[11,12]
        df['NRY5-10'] = dfinput.iat[12,12]
        df['NRY11-17'] = dfinput.iat[13,12]
        df['NRYG18'] = dfinput.iat[14,12]
        df['Total_Non-responder'] = df['NRML6']+df['NRM6-59']+df['NRY5-10']+df['NRY11-17']+df['NRYG18']

        df['MTML6'] = dfinput.iat[10,12]
        df['MTM6-59'] = dfinput.iat[11,12]
        df['MTY5-10'] = dfinput.iat[12,12]
        df['MTY11-17'] = dfinput.iat[13,12]
        df['MTYG18'] = dfinput.iat[14,12]
        df['Total_Medical_Transfer'] = df['MTML6']+df['MTM6-59']+df['MTY5-10']+df['MTY11-17']+df['MTYG18']

        df['TOPML6'] = dfinput.iat[10,13]
        df['TOPM6-59'] = dfinput.iat[11,13]
        df['TOPY5-10'] = dfinput.iat[12,13]
        df['TOPY11-17'] = dfinput.iat[13,13]
        df['TOPYG18'] = dfinput.iat[14,13]
        df['Total_Out_to_OTP'] = df['TOPML6']+df['TOPM6-59']+df['TOPY5-10']+df['TOPY11-17']+df['TOPYG18']

        df['TIPML6'] = dfinput.iat[10,14]
        df['TIPM6-59'] = dfinput.iat[11,14]
        df['TIPY5-10'] = dfinput.iat[12,14]
        df['TIPY11-17'] = dfinput.iat[13,14]
        df['TIPYG18'] = dfinput.iat[14,14]
        df['Total_Out_to_IPD'] = df['TIPML6']+df['TIPM6-59']+df['TIPY5-10']+df['TIPY11-17']+df['TIPYG18']


        df['DISML6'] = df['CURML6']+df['DIEML6']+df['UNKML6']+df['DEFML6']+df['NRML6']+df['MTML6']+df['TOPML6']+df['TIPML6']
        df['DISM6-59'] = df['CURM6-59']+df['DIEM6-59']+df['UNKM6-59']+df['DEFM6-59']+df['NRM6-59']+df['MTM6-59']+df['TOPM6-59']+df['TIPM6-59']
        df['DISY5-10'] = df['CURY5-10']+df['DIEY5-10']+df['UNKY5-10']+df['DEFY5-10']+df['NRY5-10']+df['MTY5-10']+df['TOPY5-10']+df['TIPY5-10']
        df['DISY11-17'] =df['CURY11-17']+df['DIEY11-17']+df['UNKY11-17']+df['DEFY11-17']+df['NRY11-17']+df['MTY11-17']+df['TOPY11-17']+df['TIPY11-17']
        df['DISYG18'] = df['CURYG18']+df['DIEYG18']+df['UNKYG18']+df['DEFYG18']+df['NRYG18']+df['MTYG18']+df['TOPYG18']+df['TIPYG18']
        df['Total_Discharge'] = df['DISML6']+df['DISM6-59']+df['DISY5-10']+df['DISY11-17']+df['DISYG18']

        df['ENDML6'] = df['BEGML6']+df['TADMML6']-df['DISML6']
        df['ENDM6-59'] = df['BEGM6-69']+df['TADMM6-59']-df['DISM6-59']
        df['ENDY5-10'] = df['BEGY5-10']+df['TADMY5-10']-df['DISY5-10']
        df['ENDY11-17'] = df['BEGY11-17']+df['TADMY11-17']-df['DISY11-17']
        df['ENDYG18'] = df['BEGYG18']+df['TADMYG18']-df['DISYG18']
        df['Total_End_of_Month'] = df['Total_Beginning_of_Month']+df['Total_Admission']-df['Total_Discharge']

        df.dropna()
        df.to_csv('output.csv', mode='a', header=False)
```
